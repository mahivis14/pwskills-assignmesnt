```python
class BankAccount:
    def __init__(self, account_number, account_holder, balance=0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposit of ${amount} successful. New balance: ${self.balance}")
        else:
            print("Invalid deposit amount. Please enter a positive amount.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawal of ${amount} successful. New balance: ${self.balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds.")

    def display_balance(self):
        print(f"Account Balance for {self.account_holder}: ${self.balance}")



account1 = BankAccount(account_number="123456789", account_holder="John Doe", balance=1000)

account1.display_balance()  # Display initial balance

account1.deposit(500)  # Deposit $500
account1.withdraw(200)  # Withdraw $200
account1.display_balance()  # Display updated balance
```

```python
class Employee:
    def __init__(self, employee_id, name, salary):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary

    def calculate_yearly_bonus(self, bonus_percentage):
        if bonus_percentage < 0:
            print("Invalid bonus percentage. Please enter a non-negative value.")
            return 0

        bonus = (bonus_percentage / 100) * self.salary
        return bonus

    def display_employee_details(self):
        print(f"Employee ID: {self.employee_id}")
        print(f"Name: {self.name}")
        print(f"Salary: ${self.salary}")
        print("")

employee1 = Employee(employee_id="E123", name="Alice", salary=50000)

employee1.display_employee_details()  # Display initial employee details

bonus_percentage = 10
yearly_bonus = employee1.calculate_yearly_bonus(bonus_percentage)
print(f"Yearly Bonus ({bonus_percentage}%): ${yearly_bonus}")

employee1.display_employee_details()  # Display updated employee details with bonus
```

```python
class VehicleRentalSystem:
    def __init__(self):
        self.available_vehicles = {'Car': 10, 'Motorcycle': 5, 'Truck': 3}
        self.rented_vehicles = {'Car': 0, 'Motorcycle': 0, 'Truck': 0}

    def rent_vehicle(self, vehicle_type, quantity=1):
        if vehicle_type in self.available_vehicles and quantity > 0:
            if self.available_vehicles[vehicle_type] >= quantity:
                self.available_vehicles[vehicle_type] -= quantity
                self.rented_vehicles[vehicle_type] += quantity
                print(f"Rented {quantity} {vehicle_type}(s).")
            else:
                print(f"Sorry, not enough available {vehicle_type}s.")
        else:
            print("Invalid vehicle type or quantity.")

    def return_vehicle(self, vehicle_type, quantity=1):
        if vehicle_type in self.rented_vehicles and quantity > 0:
            if self.rented_vehicles[vehicle_type] >= quantity:
                self.available_vehicles[vehicle_type] += quantity
                self.rented_vehicles[vehicle_type] -= quantity
                print(f"Returned {quantity} {vehicle_type}(s).")
            else:
                print(f"Invalid quantity to return for {vehicle_type}.")
        else:
            print("Invalid vehicle type or quantity.")

    def display_available_vehicles(self):
        print("Available Vehicles:")
        for vehicle_type, quantity in self.available_vehicles.items():
            print(f"{vehicle_type}s: {quantity}")
            
rental_system = VehicleRentalSystem()

rental_system.display_available_vehicles()  # Display initial available vehicles

rental_system.rent_vehicle('Car', 2)  # Rent 2 cars
rental_system.rent_vehicle('Motorcycle', 1)  # Rent 1 motorcycle

rental_system.display_available_vehicles()  # Display updated available vehicles

rental_system.return_vehicle('Car', 1)  # Return 1 car
rental_system.return_vehicle('Truck', 2)  # Attempt to return 2 trucks (not rented)

rental_system.display_available_vehicles()  # Display final available vehicles
```

```python
class Book:
    def __init__(self, title, author, isbn):
        """Initialize a Book object."""
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_borrowed = False

    def __str__(self):
        """Return a string representation of the book."""
        return f"{self.title} by {self.author} (ISBN: {self.isbn})"

    def borrow_book(self):
        """Borrow the book."""
        if not self.is_borrowed:
            self.is_borrowed = True
            return True
        else:
            return False

    def return_book(self):
        """Return the book."""
        if self.is_borrowed:
            self.is_borrowed = False
            return True
        else:
            return False


class Library:
    def __init__(self):
        """Initialize a Library object."""
        self.books = []

    def add_book(self, book):
        """Add a book to the library."""
        self.books.append(book)
        print(f"Book added to the library: {book}")

    def borrow_book(self, title):
        """Borrow a book from the library."""
        for book in self.books:
            if book.title == title and not book.is_borrowed:
                if book.borrow_book():
                    print(f"Book borrowed successfully: {book}")
                    return True
                else:
                    print(f"Book is already borrowed: {book}")
                    return False
        print(f"Book not found or already borrowed: {title}")
        return False

    def return_book(self, title):
        """Return a borrowed book to the library."""
        for book in self.books:
            if book.title == title and book.is_borrowed:
                if book.return_book():
                    print(f"Book returned successfully: {book}")
                    return True
                else:
                    print(f"Book is not currently borrowed: {book}")
                    return False
        print(f"Book not found or not currently borrowed: {title}")
        return False

    def display_available_books(self):
        """Display the currently available books in the library."""
        if not self.books:
            print("No books available in the library.")
        else:
            print("Available Books:")
            for book in self.books:
                if not book.is_borrowed:
                    print(book)

# Example usage:
library = Library()

book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", "9780140283334")
book2 = Book("To Kill a Mockingbird", "Harper Lee", "9780061120084")
book3 = Book("1984", "George Orwell", "9780451524935")

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

library.display_available_books()  # Display initial available books

library.borrow_book("The Great Gatsby")  # Borrow a book
library.borrow_book("The Great Gatsby")  # Attempt to borrow the same book again

library.display_available_books()  # Display updated available books

library.return_book("The Great Gatsby")  # Return the borrowed book
library.return_book("The Great Gatsby")  # Attempt to return the same book again

library.display_available_books()  # Display final available books
```

```python
class Product:
    def __init__(self, product_id, name, price, quantity):
        """Initialize a Product object."""
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity

    def __str__(self):
        """Return a string representation of the product."""
        return f"{self.name} (ID: {self.product_id}) - Price: ${self.price}, Quantity: {self.quantity}"

class InventorySystem:
    def __init__(self):
        """Initialize an InventorySystem object."""
        self.products = []

    def add_product(self, product):
        """Add a product to the inventory."""
        self.products.append(product)
        print(f"Product added to the inventory: {product}")

    def update_quantity(self, product_id, new_quantity):
        """Update the quantity of a product in the inventory."""
        for product in self.products:
            if product.product_id == product_id:
                product.quantity = new_quantity
                print(f"Quantity updated for {product}: {new_quantity}")
                return True
        print(f"Product not found: {product_id}")
        return False

    def display_available_products(self):
        """Display the currently available products in the inventory."""
        if not self.products:
            print("No products available in the inventory.")
        else:
            print("Available Products:")
            for product in self.products:
                print(product)

inventory_system = InventorySystem()

product1 = Product(product_id="P001", name="Laptop", price=999.99, quantity=10)
product2 = Product(product_id="P002", name="Smartphone", price=599.99, quantity=20)
product3 = Product(product_id="P003", name="Tablet", price=299.99, quantity=15)

inventory_system.add_product(product1)
inventory_system.add_product(product2)
inventory_system.add_product(product3)

inventory_system.display_available_products()  # Display initial available products

inventory_system.update_quantity("P001", 5)  # Update quantity for a product
inventory_system.update_quantity("P004", 8)  # Attempt to update quantity for a non-existent product

inventory_system.display_available_products()  # Display final available products
```

```python
class Shape:
    def __init__(self, length=0, width=0, height=0):
        self.length = length
        self.width = width
        self.height = height

    def calculate_area(self):
        """Calculate and return the area of the shape."""
        raise NotImplementedError("Subclasses must implement this method.")

    def calculate_perimeter(self):
        """Calculate and return the perimeter of the shape."""
        raise NotImplementedError("Subclasses must implement this method.")

class Rectangle(Shape):
    def calculate_area(self):
        """Calculate and return the area of the rectangle."""
        return self.length * self.width

    def calculate_perimeter(self):
        """Calculate and return the perimeter of the rectangle."""
        return 2 * (self.length + self.width)

class Triangle(Shape):
    def calculate_area(self):
        """Calculate and return the area of the triangle."""
        return 0.5 * self.length * self.height

    def calculate_perimeter(self):
        """Calculate and return the perimeter of the triangle."""
        return self.length + self.width + self.height

# Example usage:
rectangle = Rectangle(length=5, width=3)
triangle = Triangle(length=4, width=3, height=2)

print("Rectangle:")
print(f"Area: {rectangle.calculate_area()}")
print(f"Perimeter: {rectangle.calculate_perimeter()}")
print()

print("Triangle:")
print(f"Area: {triangle.calculate_area()}")
print(f"Perimeter: {triangle.calculate_perimeter()}")
```

```python
class Student:
    def __init__(self, student_id, name, grades=[]):
        """Initialize a Student object."""
        self.student_id = student_id
        self.name = name
        self.grades = grades

    def add_grade(self, grade):
        """Add a grade to the student's list of grades."""
        self.grades.append(grade)
        print(f"Grade {grade} added for {self.name}.")

    def calculate_average_grade(self):
        """Calculate and return the average grade of the student."""
        if not self.grades:
            print(f"No grades available for {self.name}.")
            return None
        else:
            average_grade = sum(self.grades) / len(self.grades)
            return average_grade

    def display_student_details(self):
        """Display details of the student."""
        print(f"Student ID: {self.student_id}")
        print(f"Name: {self.name}")
        print("Grades:", ", ".join(map(str, self.grades)))
        average_grade = self.calculate_average_grade()
        if average_grade is not None:
            print(f"Average Grade: {average_grade:.2f}")
        print("")

# Example usage:
student1 = Student(student_id="S001", name="Alice", grades=[90, 85, 92])
student2 = Student(student_id="S002", name="Bob", grades=[78, 88, 95])

student1.display_student_details()  # Display details for student1
student2.display_student_details()  # Display details for student2

student1.add_grade(95)  # Add a grade for student1
student2.add_grade(82)  # Add a grade for student2

student1.display_student_details()  # Display updated details for student1
student2.display_student_details()  # Display updated details for student2
```

```python
class Email:
    def __init__(self, sender, recipient, subject, body=""):
        """Initialize an Email object."""
        self.sender = sender
        self.recipient = recipient
        self.subject = subject
        self.body = body

    def send_email(self):
        """Send the email."""
        print(f"Email sent from {self.sender} to {self.recipient} with subject: {self.subject}")
        print("Body:")
        print(self.body)
        print("Email sent successfully.")

    def display_email_details(self):
        """Display details of the email."""
        print("Email Details:")
        print(f"Sender: {self.sender}")
        print(f"Recipient: {self.recipient}")
        print(f"Subject: {self.subject}")
        print("Body:")
        print(self.body)
        print("")

# Example usage:
email1 = Email(sender="alice@example.com", recipient="bob@example.com", subject="Meeting Invitation", body="Dear Bob, ...")
email2 = Email(sender="charlie@example.com", recipient="alice@example.com", subject="Project Update", body="Hi Alice, ...")

email1.display_email_details()  # Display details for email1
email2.display_email_details()  # Display details for email2

email1.send_email()  # Send email1
email2.send_email()  # Send email2
```

```python
class SocialMediaProfile:
    def __init__(self, username):
        """Initialize a SocialMediaProfile object."""
        self.username = username
        self.posts = []

    def add_post(self, post_text):
        """Add a post to the social media profile."""
        self.posts.append(post_text)
        print(f"New post added to {self.username}'s profile: {post_text}")

    def display_posts(self):
        """Display all posts in the social media profile."""
        if not self.posts:
            print(f"No posts available for {self.username}.")
        else:
            print(f"Posts by {self.username}:")
            for i, post in enumerate(self.posts, start=1):
                print(f"{i}. {post}")
            print("")

    def search_posts(self, keyword):
        """Search for posts containing a specific keyword."""
        matching_posts = [post for post in self.posts if keyword.lower() in post.lower()]
        if not matching_posts:
            print(f"No posts found containing the keyword '{keyword}' for {self.username}.")
        else:
            print(f"Posts containing the keyword '{keyword}' for {self.username}:")
            for i, post in enumerate(matching_posts, start=1):
                print(f"{i}. {post}")
            print("")

# Example usage:
profile1 = SocialMediaProfile(username="Alice")
profile2 = SocialMediaProfile(username="Bob")

profile1.add_post("Enjoying a sunny day at the park!")
profile1.add_post("Trying out a new recipe for dinner.")
profile2.add_post("Coding is so much fun!")

profile1.display_posts()  # Display posts for profile1
profile2.display_posts()  # Display posts for profile2

profile1.search_posts("recipe")  # Search for posts containing the keyword 'recipe' for profile1
profile2.search_posts("coding")  # Search for posts containing the keyword 'coding' for profile2
```

```python
from datetime import datetime

class ToDoList:
    def __init__(self):
        """Initialize a ToDoList object."""
        self.tasks = {}

    def add_task(self, task, due_date=None):
        """Add a task to the ToDo list with an optional due date."""
        if due_date is not None and not self._is_valid_due_date(due_date):
            print("Invalid due date format. Please use 'YYYY-MM-DD'.")
            return

        self.tasks[task] = {'due_date': due_date, 'completed': False}
        print(f"Task added to the ToDo list: {task}")

    def mark_task_completed(self, task):
        """Mark a task as completed."""
        if task in self.tasks:
            self.tasks[task]['completed'] = True
            print(f"Task marked as completed: {task}")
        else:
            print(f"Task not found: {task}")

    def display_pending_tasks(self):
        """Display pending tasks in the ToDo list."""
        pending_tasks = [task for task, details in self.tasks.items() if not details['completed']]
        if not pending_tasks:
            print("No pending tasks.")
        else:
            print("Pending Tasks:")
            for i, task in enumerate(pending_tasks, start=1):
                due_date = self.tasks[task]['due_date']
                print(f"{i}. {task} (Due Date: {due_date if due_date else 'Not specified'})")
            print("")

    def _is_valid_due_date(self, due_date):
        """Check if the due date has a valid format 'YYYY-MM-DD'."""
        try:
            datetime.strptime(due_date, '%Y-%m-%d')
            return True
        except ValueError:
            return False

# Example usage:
todo_list = ToDoList()

todo_list.add_task("Complete project", due_date="2023-01-15")
todo_list.add_task("Read a book")
todo_list.add_task("Exercise", due_date="2023-01-20")

todo_list.display_pending_tasks()  # Display pending tasks

todo_list.mark_task_completed("Read a book")  # Mark a task as completed

todo_list.display_pending_tasks()  # Display updated pending tasks
```
